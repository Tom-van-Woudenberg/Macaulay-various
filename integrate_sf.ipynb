```python
import sympy as sp

x = sp.Symbol('x',real=True)

q_z = -5*sp.SingularityFunction(x,0,-1) + 10 * sp.SingularityFunction(x,5,-1) - 5 * sp.SingularityFunction(x,10,-1)
theta = sp.pi/18
dvdx = -q_z*sp.cos(theta)
display(dvdx)
V = sp.integrate(dvdx,x)
display(V)

M = sp.integrate(V,x)*sp.cos(theta)
display(M)
```

```python
sp.plot(sp.cos(theta),(x,0,10))
```

```python
sp.plot(theta,(x,0,10))
theta_piecewise = theta.rewrite(sp.Piecewise).simplify()
display(theta_piecewise)
x_c = sp.integrate(theta_piecewise,(x))
display(x_c)
sp.plot(x_c,(x,0,10))
```
