```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sympy.physics.continuum_mechanics.beam import Beam
sym.init_printing()
```

```python
E = sym.Symbol('E')
I = sym.Symbol('I')
b = Beam(10, E, I)
R_0, R_5, R_10 = sym.symbols('R_0, R_5, R_10')
b.apply_support(0, type='pin')
b.apply_support(5, type='pin')
b.apply_support(10, type='pin')
#b.solve_for_reaction_loads(R_0, R_10)
#b.reaction_loads
b.solve_for_ild_reactions(1, R_0, R_5, R_10)
display(b.ild_reactions)
b.plot_ild_reactions();
```

```python
b.solve_for_ild_moment(4, 1, R_0, R_5, R_10)
display(b.ild_moment)
b.plot_ild_moment();
```

```python
b.solve_for_ild_shear(4, 1, R_0, R_5, R_10)
display(b.ild_shear)
b.plot_ild_shear();
```
